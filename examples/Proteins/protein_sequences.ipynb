# Execute protein sequences from a FASTA file

```python
from pprint import pprint
import requests, io
%run ../../chemw/mw.py

# examining a FASTA file from the Codons PyPI module
protein = Proteins()
protein.mass(fasta_link = 'https://github.com/freiburgermsu/codons/raw/main/examples/MERS/central_dogma/MERS_proteins.fasta')

print('\n\nresultant calculations', '\n', '='*30)
pprint(protein.fasta_protein_masses)
with open('protein.fasta', 'w') as out:
    out.write(protein.fasta)
```

### Comparison with BioPython

```python
from Bio.SeqUtils import molecular_weight
%run ../../chemw/mw.py

for seq in protein.fasta_protein_masses:
    bio_mass = molecular_weight(seq, 'protein')
    print(f'The ChemW mass of {protein.fasta_protein_masses[seq]} versus the BioPython mass of {bio_mass}.\n')
```
